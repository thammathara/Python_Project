```python
import requests 
import time
import pandas as pd 
```

#Check Status code 200

```python
test_url = "https://swapi.dev/api/species/1"

response = requests.get(test_url)

response.status_code
```

```python
name = []
classification = []
designation = []
average_height = []
skin_colors = []
hair_colors = []
eye_colors = []
average_lifespan = []
language = []

for i in range(1,37):
    url = f"https://swapi.dev/api/species/{i}"
    resp = requests.get(url)
    result = resp.json()
    name.append(result['name'])
    classification.append(result['classification'])
    designation.append(result['designation'])
    average_height.append(result['average_height'])
    skin_colors.append(result['skin_colors'])
    hair_colors.append(result['hair_colors'])
    eye_colors.append(result['eye_colors'])
    average_lifespan.append(result['average_lifespan'])
    language.append(result['language'])

    time.sleep(1)

df_species = pd.DataFrame({
    "name": name, 
    "classification": classification, 
    "designation": designation,
    "average_height": average_height,
    "skin_colors": skin_colors,
    "hair_colors": hair_colors,
    "eye_colors": eye_colors,
    "average_lifespan": average_lifespan,
    "language": language,})

df_species.tail(10)
```

# Top 10 average hightest species in Star Wars ðŸ‘½

```python
#filter out unknown & n/a in average_height
result = df_species.query(" average_height!= ['unknown','n/a']")
```

```python
#Convert String to Int
result['average_height'] = result['average_height'].astype(int)
```

```python
result.groupby(['classification','name'])['average_height'].sum().sort_values(ascending=False).head(10).reset_index()
```

#How many unique classification of species in Star wars?

```python
result['classification'].value_counts().reset_index()
```

#Which of the top 5 species have the highest average life span based on classification? (exclude indefinite)

```python
#filter out unknown, n/a and indefinite	 in average_lifespan
result1 = df_species.query(" average_lifespan!= ['unknown','n/a','indefinite']")
```

```python
#Convert String to Int
result1['average_lifespan'] = result1['average_lifespan'].astype(int)
```

```python
result1.groupby(['classification'])['average_lifespan'].mean().sort_values(ascending=False).head(5).reset_index()

```
